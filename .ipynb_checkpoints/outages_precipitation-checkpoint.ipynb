```python
#import python data science libraries
import pandas as pd
import numpy as np
import folium
import re
import geopandas
import matplotlib.pyplot as plt
```

```python
#######
# read in the data
######
outages = pd.read_csv('outwxdata_cleaned.csv')
outages.head(90)
```

```python
#drop "OutTime" column since it is no longer needed in this context
x=outages.drop(['OutTime'], axis=1)
x.head(21)
```

```python
#getting the highest value of precipitation the dataset
y=x[x["Precip"] > 1]
y.head(21)
column = y["Precip"]
max_value = column.max()
print (max_value)
```

```python
##using folium to generate a heatmap visualization 
#for a concentration of power outages caused by precipitation across the country

from folium.plugins import HeatMap

fig7=folium.Figure(width=850,height=550)
m7=folium.Map(location=[-1.296457, 36.818380],zoom_start=7)
fig7.add_child(m7)

# convert to (n, 2) nd-array format for heatmap
precips = y[['Lat', 'Long']].to_numpy()

# plot heatmap
m7.add_children(HeatMap(precips, radius=15))
m7
```

```python
#saving the visualization as an html for use in the UI
m7.save('outages_precip.html')
```
