```python
import pandas as pd
import numpy as np
import folium
import re
import geopandas
import matplotlib.pyplot as plt
```

```python
#######
# read in the data
######
outages = pd.read_csv('outwxdata_cleaned.csv')
outages.head(90)
```

```python
x=outages.drop(['OutTime'], axis=1)
x.head(21)
```

```python
y=x[x["WindGust"] > 5]
y.head(21)
column = y["WindGust"]
max_value = column.max()
print (max_value)
```

```python
from folium.plugins import HeatMap

fig7=folium.Figure(width=850,height=550)
m7=folium.Map(location=[-1.296457, 36.818380],zoom_start=7)
fig7.add_child(m7)

# convert to (n, 2) nd-array format for heatmap
windgusts = y[['Lat', 'Long']].to_numpy()

# plot heatmap
m7.add_children(HeatMap(windgusts, radius=15))
m7
```

```python
m7.save('outages_wind.html')
```
