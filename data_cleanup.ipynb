```python
import pandas as pd

#load the dataset
dataset = pd.read_csv('outwxdata.csv')

#show the first 5 rows of the dataset
dataset.tail(5)
```

```python
dataset = dataset[['LocName', 'Lat', 'Long', 'OutTime', 'Temp', 'Precip', 'WindSpd', 'WindGust', 'Press', 'Humid', 'Dew']]

dataset.head(5)
```

```python
import numpy as np

holiday = pd.read_excel('kenya-holidays.xlsx')

#convert the datetime format to just 'date'
holiday_dates = np.array(holiday['Dates'].dt.date)

#adding a columns of zeros for 'Holiday'
dataset['Holiday'] = np.zeros((dataset.shape[0]))
```

```python
dataset['OutTime'] = pd.to_datetime(dataset['OutTime'])

#extracting only the dates
outage_dates = dataset['Date'] = np.array(dataset['OutTime'].dt.date)

#adding holiday data
#iterating through all the holiday dates
for h_d in holiday_dates:   
    #if an outage date is found to coincide with a holiday date
    if h_d in outage_dates:
        #create a list of indicies where the dates match
        lt = np.where(outage_dates == h_d)
        #go through the list and mark the 'Holiday' column as '1'
        for i in lt:
            dataset['Holiday'][i] = 1

#finally we count the number of dates which where holidays vs not holidays            
dataset.groupby('Holiday').count()
```

```python
dataset = dataset[['Lat', 'Long', 'OutTime', 'Temp', 'Precip', 'WindSpd', 'WindGust', 'Press', 'Humid', 'Dew', 'Holiday']]
dataset.head(10)
```
