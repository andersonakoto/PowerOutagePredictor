```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('outwxdata_fine.csv')
df.head()
```

```python
to_drop = ['OutID',
           'LocName']

df.drop(to_drop, inplace=True, axis=1)
```

```python
df.head()
```

```python
df = df[['Lat','Lng','OutTime','Temp','Precip','WindSpd', 'WindGust', 'Press', 'Humid', 'Dew', 'Type']]
df.head()
```

```python
df['Outage'] = np.where(df['Type']== "Unplanned", '1', '0')
df.head()
```

```python
df.tail()
```

```python
to_drop = ['Type']

df.drop(to_drop, inplace=True, axis=1)

df.head()
```
